```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('/home/siva/Downloads/py-master/ML/13_kmeans/income.csv')
df
```

```python
df.drop(['Name'],axis=1,inplace=True)
df
```

```python
scalar=MinMaxScaler()
df['Age']=scalar.fit_transform(df[['Age']])
df
```

```python
df['Income($)']=scalar.fit_transform(df[['Income($)']])
df
```

```python
sse=np.array([])
```

```python
from sklearn.cluster import KMeans
```

```python
for i in range(1,len(df['Age'])):
    model=KMeans(n_clusters=i)
    model.fit(df)
    sse=np.append(sse,model.inertia_)
```

```python
sse
```

```python
sse_dict={x:y for x,y in enumerate(sse)}
sse_dict
```

```python
sns.scatterplot(x=sse_dict.keys(),y=sse_dict.values())
```

Therefore by the elbow method it is evident that the ideal number of clusters is 3!

```python
model=KMeans(n_clusters=3)
kmeans=model.fit(df[["Age",'Income($)']])
df['clusters']=kmeans.labels_
df
```

```python
sns.scatterplot(x=kmeans.cluster_centers_[:,0],\
                y=kmeans.cluster_centers_[:,1])
sns.scatterplot(x="Age",y='Income($)',data=df,hue='clusters')
```
